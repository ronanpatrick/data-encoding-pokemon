```python
import pandas as pd
from google.colab import drive
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('pokemon (1).csv')
df
```

```python
df1 = df[['type1' , 'generation']]
df1
```

```python
# Ordinal
df1['generation'].unique()
```

```python
# Ordinal
gen_order = ['1' , '2' , '3' , '4' , '5' , '6' , '7' , '8']
enc = OrdinalEncoder(categories = [gen_order])
```

```python
# Ordinal
enc.fit_transform(df1[['generation']])
```

```python
# Ordinal
df1['generation'] = enc.fit_transform(df1[['generation']])
df1
```

```python
# One Hot Encoder
ohe = OneHotEncoder(handle_unknown = 'ignore' , sparse_output = False).set_output(transform = 'pandas')
```

```python
df1['type1'].unique()
```

```python
ohe_transform = ohe.fit_transform(df1[['type1']])
ohe_transform
```

```python
df1 = pd.concat([df1 , ohe_transform] , axis = 1)
display(df1, ohe_transform)
```

```python
df1.drop(columns = ['type1'] , inplace = True)
df1
```
